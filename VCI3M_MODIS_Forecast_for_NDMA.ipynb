<b>This notebook demonstrates the use of the AR method to forecast the 3 monthly VCI.</b>

<b>Forecasts can be done for any of 29 county livelihood zone intersections.</b>

<b>First we load the data.</b>

```python
region = plot_func_adam.which_region()
```

```python
T,VCI1W,VCI3M = plot_func_adam.load(region)
```

<b>Then we plot the data:</b>

```python
plot_func_adam.plot_vci(T,VCI3M,'3 month VCI')                   
```

<b>The solid black lines here show VCI=35, the threshold below which the region is said to be in drought.</b>

<b> Next we generate the forecast for the next 8 weeks, using the AR modelling method (and a 1 standard deviation uncertainty).</b> 

<b> We plot this, with error bars showing the plus/minus one standard deviation uncertainty. </b>

<b> And we give a forecast for 4 weeks from our last observation, which was on 15th July.</b>

```python
f,s=plot_func_adam.forecast(VCI3M)
plot_func_adam.plot_vci_fc(T,VCI3M,f,s,'3 month VCI')
```
