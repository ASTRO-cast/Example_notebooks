# Using linear autoregression (AR) to forecast VCI on MODIS data

```python
import plot_func_adam
```

<b>This notebook demonstrates the use of the AR method to forecast the weekly and 3 monthly VCI.</b>

<b>This example is for the Baringo county LZ24 intersection.</b>

<b>First we load the weekly and 3 monthly VCI data.</b>

```python
# The input here should be an array, with the first row being the dates of observation in the format of days since
# 1/1/2000, the second row being weekly VCI, and the third row being 3 month VCI.
# T = days since 1/1/2000
# VCI1W = weekly VCI, 
# VCI3M = 3 month VCI

T,VCI1W,VCI3M = plot_func_adam.load('Baringo_Z24_AR_MODIS.npy')
```

<b>Then we plot the data:</b>

```python
# The function 'plot_vci' takes the time and VCI data and VCI version as inputs and plots the data
plot_func_adam.plot_vci(T,VCI1W,'Weekly VCI')                   
plot_func_adam.plot_vci(T,VCI3M,'3 month VCI')                   
```

<b>The solid black lines here show VCI=35, the threshold below which the region is said to be in drought.</b>

<b> Next we generate a forecast for the weekly VCI for the next 8 weeks, using the AR modelling method (and a 1 standard deviation uncertainty).</b> 

```python
f,s=plot_func_adam.forecast(VCI1W)
```

<b> And we plot this, with error bars showing the plus/minus one standard deviation uncertainty: </b>

```python
# The function 'plot_vci_fc' takes the time and VCI data, the forecast and uncertainty, 
# and VCI version, as inputs and makes a plot.

plot_func_adam.plot_vci_fc(T,VCI1W,f,s,'Weekly VCI')
```

<b>We repeat this for the 3 month VCI:</b>

```python
f,s=plot_func_adam.forecast(VCI3M)
plot_func_adam.plot_vci_fc(T,VCI3M,f,s,'3 month VCI')
```

<b>To demonstrate forecasting for another region, we now show a plot of 3 month VCI for the intersection of Isiolo and LZ10:</b>

```python
T,VCI1W,VCI3M = plot_func_adam.load('Isiolo_Z10_AR_MODIS.npy')
f,s=plot_func_adam.forecast(VCI3M)
plot_func_adam.plot_vci_fc(T,VCI3M,f,s,'3 month VCI')
```
